```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
pokeman = pd.read_csv("pokemon.csv")
```

```python
pokeman.head()
```

```python
pokeman.info
```

```python
pokeman.describe().T
```

```python
pkmn_types = pokeman.melt(id_vars = ['id', 'species'], value_vars = ['type_1', 'type_2'], var_name = 'type_level', value_name = 'type')
pkmn_types.head(10)
```

```python
type_counts = pkmn_types['type'].value_counts()
```

```python
type_order = type_counts.index
```

```python
base_color = sns.color_palette()[0]
sns.countplot(data = pkmn_types, y = 'type', color = base_color, order = type_order);
```

```python
n_pokeman = pkmn_types['type'].value_counts().sum()
max_type_count = type_counts[0]
max_prop = max_type_count / n_pokeman
print(max_prop)
```

```python
tick_props = np.arange(0, max_prop, 0.02)
tick_props
```
